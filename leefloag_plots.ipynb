```python
import daft
from matplotlib import rc
from matplotlib import pyplot as plt



```

```python

rc('font', family='serif', size=12)
rc('text', usetex=True)

pgmlbl = daft.PGM()

pgmlbl.add_node('Input', r'$x$', 0, 5, observed=True)
pgmlbl.add_node('Hidden1', r'$h$', 0, 4)
pgmlbl.add_node('Prev', r'$l_{t-n} \dots l_{t-1}$', 2, 5, aspect=3.5, observed=True)
pgmlbl.add_node('Hidden2', r'$h$', 2, 4)
pgmlbl.add_node('Concat', r'$c$', 1, 3)
pgmlbl.add_node('Hidden3', r'$h$', 1, 2)
pgmlbl.add_node('Next', r'$l_t$', 1, 1, alternate=True)

pgmlbl.add_edge('Input', 'Hidden1')
pgmlbl.add_edge('Prev', 'Hidden2')
pgmlbl.add_edge('Hidden1', 'Concat')
pgmlbl.add_edge('Hidden2', 'Concat')
pgmlbl.add_edge('Concat', 'Hidden3')
pgmlbl.add_edge('Hidden3', 'Next')

pgmlbl.render()
pgmlbl.savefig('lbl.png')
```

```python
aoo = daft.PGM()

aoo.add_node('Input', r'$x$', 0, 5, observed=True)
aoo.add_node('Hidden1', r'$h^1$', 0, 4)
aoo.add_node('Hidden2l', r'$h^2_1$', -1, 3)
aoo.add_node('Hidden2r', r'$h^2_n$', 1, 3)

aoo.add_node('b1', r'', -.2, 3, fixed=True)
aoo.add_node('b2', r'', 0, 3, fixed=True)
aoo.add_node('b3', r'', .2, 3, fixed=True)

aoo.add_node('b4', r'', -.2, 2, fixed=True)
aoo.add_node('b5', r'', 0, 2, fixed=True)
aoo.add_node('b6', r'', .2, 2, fixed=True)

aoo.add_node('Outl', r'$y_1$', -1, 2, alternate=True)
aoo.add_node('Outr', r'$y_n$', 1, 2, alternate=True)

aoo.add_edge('Input', 'Hidden1')
aoo.add_edge('Hidden1', 'Hidden2l')
aoo.add_edge('Hidden1', 'Hidden2r')
aoo.add_edge('Hidden2l', 'Outl')
aoo.add_edge('Hidden2r', 'Outr')


aoo.render()
aoo.savefig('aoo.png')
```
